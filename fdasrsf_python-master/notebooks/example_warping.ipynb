# Function Example
This notebook will show how to use the `fdasrsf` package to align and statistically analyze a set of functions using the SRVF framework

## Load Packages

```python
import fdasrsf as fs
import numpy as np
import matplotlib.pyplot as plt
data = np.load('../bin/simu_data.npz')
time = data['arr_1']
f = data['arr_0']
```

## Analyze
We now will construct a `fdawarp` object

```python
obj = fs.fdawarp(f,time)
```

Will now align the functions

```python
obj.srsf_align(parallel=True)
```

Plot the results

```python
obj.plot()
```

## Vertical PCA
Now analyze the amplitude component using vertical fPCA

```python
vpca = fs.fdavpca(obj)
vpca.calc_fpca(no=10)
vpca.plot()
```

## Horizontal PCA
Now analyze the phase component using horizontal fPCA

```python
hpca = fs.fdahpca(obj)
hpca.calc_fpca(no=3)
hpca.plot()
```

## Joint fPCA
Sometimes the data has correlation and we can analyze the amplitude and phase jointly using joint fPCA

```python
jpca = fs.fdajpca(obj)
jpca.calc_fpca(no=10)
jpca.plot()
```
